## The Galilean Model

- Formulated polynomials that capture accelaration, speed and distance. Where Galileos model says the horizontal position of a thrown ball at time $t$ will be given by:
$$x=v_1 t$$
- Where $v_1$ represents the starting speed of the ball in the $x$ (horizontal) direction. The height of a thrown ball $(y)$, according to galileo , can be calculated at time $t$ as:
$$y=v_2 t+ \dfrac{at^2}{2}$$

- Where $v_2$ represent the starting speed of the ball in the vertical direction and $a$ represents the constant downward acceleration due to gravity. When we substitute the first equation into the second equation, we find that the height of a thrown ball $y$ relates to the horizontal position of the ball $x$ as follows:
$$y = \dfrac{v_2}{v_1}x+\dfrac{ax^2}{2v^2_1}$$


- We can use Galileo's equations to model a hypothetical ball's trajectory in Python as follows:
```python
def ball_trajectory(x):
    location = 10*x - 5*(x**2)
    return location
```

````python
# Lets plot the output of a hypothetical scenario

import matplotlib.pyplot as plt

def ball_trajectory(x):
    location = 10*x - 5*(x**2)
    return location

xs = [x/100 for x in list(range(200))]
ys = [ball_trajectory(x) for x in xs]
plt.plot(xs, ys)
plt.title('The Trajectory of a thrown ball')
plt.xlabel('Horizontal Position of ball')
plt.ylabel('Vertical Position of ball')
plt.axhline(y = 0)
plt.show()

````
